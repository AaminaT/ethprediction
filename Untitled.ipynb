```python
import json, glob
from pathlib import Path

root   = Path("/Users/aaminathasneemkhaleel/Downloads/170-171-new")
paths  = sorted(glob.glob(str(root / "blockTransactions*.json")))

blocks_rows = []        
tx_rows     = []         

for fp in paths:
    with open(fp) as f:
        # each file is a top-level list:  [ [block, tx1, tx2, …],  [block, tx1, …], … ]
        outer_list = json.load(f)

    for row in outer_list:
        if not row:                    # defensive
            continue

        block_dict = row[0]            # index 0  → block-level info
        blocks_rows.append(block_dict)

        for tx in row[1:]:             # rest     → transaction dicts
            tx["blockHash"] = block_dict["blockHash"]   # keep the join key
            tx_rows.append(tx)

blocks_df = pd.json_normalize(blocks_rows)
tx_df     = pd.json_normalize(tx_rows)

print("Blocks:", blocks_df.shape)
print("Txs:   ", tx_df.shape)
```

```python
from itertools import chain
root   = Path("/Users/aaminathasneemkhaleel/Downloads/170-171-new")
files  = [
    root / "blockTransactions17010001-17011000.json",
    root / "blockTransactions17011001-17012000.json",
    root / "blockTransactions17012001-17015000.json",
    root / "blockTransactions17015001-17020000.json",
    root / "blockTransactions17020001-17030000.json",
    root / "blockTransactions17030001-17050000.json",
    root / "blockTransactions17090001-17100000.json",
]
#root / "blockTransactions17012001-17015000.json",
#     root / "blockTransactions17015001-17020000.json",
#     root / "blockTransactions17020001-17030000.json",
#     root / "blockTransactions17030001-17050000.json",
#     root / "blockTransactions17090001-17100000.json",

# ------------------------------------------------------------
# 2.  Walk through every file once, extracting rows
# ------------------------------------------------------------
block_rows, tx_rows  = [], []
union_block_keys     = set()      # we'll build the superset of keys on the fly
union_tx_keys        = set()

for fp in files:
    with open(fp) as f:
        outer = json.load(f)      # outer is:  [ [block, tx1, tx2…],  [block, tx1…], … ]

    for record in outer:
        if not record:            # guard against empty lists
            continue

        # ----- block -----
        block_dict = record[0]
        block_rows.append(block_dict)
        union_block_keys.update(block_dict)      # collect all block keys

        # ----- transactions -----
        tx_dicts = record[1:]
        for tx in tx_dicts:
            tx["blockHash"] = block_dict["hash"] if "hash" in block_dict else block_dict["blockHash"]
            tx_rows.append(tx)
        union_tx_keys.update(chain.from_iterable(tx_dicts))
```

```python
blocks_df = (
    pd.json_normalize(block_rows)
      .reindex(columns=sorted(union_block_keys))
)

tx_df = (
    pd.json_normalize(tx_rows)
      .reindex(columns=sorted(union_tx_keys))
)
```
